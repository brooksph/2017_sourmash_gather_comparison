```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")


plt.rc('font', weight='bold')
plt.rc('xtick.major', size=5, pad=7)
plt.rc('xtick', labelsize=15)

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(14, 14))

# Load the example car crash dataset
dx = pd.read_csv("../work/podar-reads.10k.sig.csv", header=0).sort_values("f_match", ascending=False)


#Plot the crashes where alcohol was involved
sns.set_color_codes("muted")
sns.barplot(x="f_match", y="name", data=dx,
            label="Genome Fraction recovered in search against genbank", color="b")

# Add a legend and informative axis label
ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(xlim=(0, 1), ylabel="",
       xlabel="")
#ax.tick_params(axis='x', which='major', pad=15)
#ax.set_title("What an Awesome Plot!")
sns.despine(left=True, bottom=True)
```

```python
da = dx.rename(columns={'f_match': 'f_match_2'})

da
```

```python
danew = da[['name','f_match_2']].copy()
danew_i = danew.set_index('name')
dxnew = df[['name','f_match']].copy()
dxnew_i = dxnew.set_index('name')
result = danew_i.join(dxnew_i, how='outer')
result
result.to_csv("out.csv")
```

```python
sns.set(style="whitegrid")


plt.rc('font', weight='bold')
plt.rc('xtick.major', size=5, pad=7)
plt.rc('xtick', labelsize=24)

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(14, 20))

x = pd.read_csv("out.csv", header = 0).sort_values("f_match", ascending=False)

# Load the example car crash dataset
#df = pd.read_csv("../work/gather_output_mircea_reads_v_known_k31_scaled_10000.csv").sort_values("f_match", ascending=False)
#dx = pd.read_csv("../work/podar-reads.10k.sig.csv").sort_values("f_match", ascending=False)

# Plot the total crashes
sns.set_color_codes("pastel")
sns.barplot(x="f_match", y="name", data=x,
            label="Genome Fraction recovered in search against known", color="b")

# Plot the crashes where alcohol was involved
sns.set_color_codes("muted")
sns.barplot(x="f_match_2", y="name", data=x,
           label="Genome Fraction recovered in search against genbank", color="b", alpha=0.5)

# Add a legend and informative axis label
ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(xlim=(0, 1), ylabel="",
       xlabel="")
#ax.tick_params(axis='x', which='major', pad=15)
#ax.set_title("What an Awesome Plot!")
sns.despine(left=True, bottom=True)
sns.set_context('poster')
```
